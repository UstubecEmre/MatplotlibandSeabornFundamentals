# 1. What is Violinplot?

A Violin Plot is a chart that shows the distribution of a numeric variable â€” similar to a box plot, but more detailed.

It combines:

A box plot (showing median, quartiles, etc.)

A density plot (showing how values are spread out)

ðŸŸ£ In Seaborn, it's created using sns.violinplot().


### 1.1. âœ… What It Looks Like

A violin plot looks like a symmetrical shape (like a violin ðŸŽ»), which shows:

Where the data is concentrated (fatter areas)

Where the data is sparse (thinner areas)

Median and quartile lines inside (just like a boxplot)

Created in Seaborn using sns.violinplot()

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset and convert DataFrame
def load_dataset(df_name, file_path):
    file_path = file_path
    df_name = df_name
    
    df_name = pd.read_csv(file_path)
    if df_name is None:
        raise FileNotFoundError("File Path Not Found")
    
    df = pd.DataFrame(df_name)
    return df


df = load_dataset(file_path= r"D:\DataVisualizationForPython\datasets\supermarket_sales_utf8.csv",df_name = "sales")

```

```python
# fix column names
def fix_column_names(dataframe):
    dataframe.columns = [col.lower().replace(" ", "_") for col in dataframe.columns]
    
# call the fix_column_names
fix_column_names(df)
df.head()
```

```python
# Violinplot
plt.figure(figsize = (5, 3))
sns.violinplot(data = df, x = "quantity", color = "orange") # horizontal
plt.xlabel("Quantity")
plt.tight_layout()
plt.show()
```

```python
# Example
sns.violinplot(data = df, x = "customer_type", y = "cogs" )
plt.title("Violinplot For Cogs Per Customer Type")
plt.tight_layout()
plt.show()
```

```python
# two categorical and one numerical variable
sns.violinplot(data = df, x = "customer_type", hue = "branch", hue_order = ["A","B","C"], y = "cogs" )
plt.title("Violinplot For Cogs Per Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("COGS")
plt.tight_layout()
plt.show()
```

```python
# split and inner arguments
sns.violinplot(data = df, 
               x = "customer_type", 
               hue = "gender", 
               hue_order = ["Male","Female"], 
               y = "cogs", 
               split = True,
               inner = "quartile",
               palette="autumn")
plt.title("Violinplot For Cogs Per Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("COGS")
plt.tight_layout()
plt.show()
```

```python
# cut and band-width argument
plt.figure(figsize = (15, 10))
sns.violinplot(data = df, 
               x = "product_line", 
               hue = "gender", 
               hue_order = ["Male","Female"], 
               y = "cogs", 
               split = True,
               inner = "stick",
               palette="viridis",
               bw = 0.4,
               cut = 0)
plt.title("Violinplot For Cogs Per Product Line")
plt.xlabel("Product Line")
plt.ylabel("COGS")
plt.tight_layout()
plt.show()
```
