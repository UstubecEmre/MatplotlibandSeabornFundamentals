# 1. What is pairplot in Seaborn?
Scatter plots for every pair of numeric columns

Histograms or KDE plots on the diagonals (showing individual distributions)

It helps visualize the pairwise relationships between multiple variables in one figure.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

```

```python
# import dataset 
def load_dataset(df_name, file_path):
    file_path = file_path
    df_name = df_name
    
    df_name = pd.read_csv(file_path)
    if df_name is None:
        raise FileNotFoundError("File Path Not Found")
    
    df = pd.DataFrame(df_name)
    return df

# assign
df = load_dataset(file_path= r"D:\DataVisualizationForPython\datasets\supermarket_sales_utf8.csv",df_name = "sales")
```

```python
# fix column names
def fix_column_names(dataframe):
    dataframe.columns = [col.lower().replace(" ", "_") for col in dataframe.columns]
    
# call the fix_column_names
fix_column_names(df)
df.head()
```

```python
num_cols = [col for col in df.columns if df[col].dtype in ['float', "int"]]
print(num_cols)
```

```python
plt.figure(figsize = (20, 15))
sns.pairplot(data = df[num_cols])
plt.title("Pair Plot of Each Numeric Column")
plt.tight_layout()
plt.show()
```

```python
# kind attribute
plt.figure(figsize = (20, 15))
sns.pairplot(data = df[num_cols], kind = 'kde', palette="BuGn")
plt.title("Pair Plot of Each Numeric Column")
plt.tight_layout()
plt.show()
```

```python
# kind = scatter | kde | hist | reg 
plt.figure(figsize = (20, 15))
sns.pairplot(data = df[num_cols], kind = 'reg', palette="Dark2")
plt.title("Pair Plot of Each Numeric Column")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data = df, x_vars = ['unit_price',"gross_income"], y_vars = ["cogs","total"], diag_kind = None, palette = "coolwarm")
```

```python
plt.figure(figsize= (20, 15))
sns.pairplot(data = df,
             diag_kws= dict(
                 color = "yellow",
                 kde = True
             ),
             plot_kws = dict(
                 color = "purple"
             ))
```
