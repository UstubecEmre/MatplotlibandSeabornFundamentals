# 1. What is a Line Plot in Seaborn?
ðŸ”¹ A line plot connects data points with straight lines.

It shows how a numeric variable changes over another variable, typically over time.

It's ideal for trend analysis.

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
file_path = r"D:\DataVisualizationForPython\datasets\orders.xlsx"

file = pd.read_excel(file_path)

if file is None:
    raise FileNotFoundError("File Path Not Found. Please, Check Your File Path")

df = pd.DataFrame(file)
df.head()
```

```python
# Basic line plot
plt.figure(figsize = (8,6))
sns.lineplot(data = df, x = "Outlet_Year", y ="Sales", ci = None)
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.tight_layout()
#plt.xticks(np.arange(0,df["Sales"].max()))
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.lineplot(data = df, x = "Outlet_Year", y ="Sales", ci = None, estimator="sum", color = "darkgray")
plt.xlabel("Year")
plt.ylabel("Sum of Sales")
plt.tight_layout()
plt.show()
```

```python
age = np.arange(20, 60, 10)
frequency = age + 15
print(age)
print(frequency)

plt.figure(figsize=(6,4))
sns.lineplot(x = age, y = frequency, color = "yellow", marker = "o")
plt.title("Linear Plot of Age Frequencies")
plt.xlabel("Age")
plt.ylabel("Frequency")
# plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(ticks = age)
plt.tight_layout()
plt.show()

```

```python
# create years
years = np.arange(2000, 2025, 1)
print(years)

# create sales
sales = np.random.normal(600, 24, 100).reshape(25,4) # mean, std, sample

# convert float to int
int_sales = np.round(sales, 0).astype(int)
print(int_sales)

df = pd.DataFrame(int_sales, columns=["Q1", "Q2", "Q3", "Q4"])
df["Year"] = years
df.head()
```

```python
df.groupby(by = "Year")[["Q1","Q3"]].agg(
    {
    "Q1": ["sum","mean"],
    "Q3":["sum","mean"]
    }
)
```

```python
sns.lineplot(data = df, x = "Year", y = "Q1", ci = None, palette = "Paired_r")
plt.title("Line Plot of Q1 Per Each Year")
plt.xlabel("Year")
plt.ylabel("Q1")
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(data = df, x = "Year", y = "Q2", ci = None, color = "red")
plt.title("Line Plot of Q2 Per Each Year")
plt.xlabel("Year")
plt.ylabel("Q2")
plt.tight_layout()
plt.show()
```

```python
# use subplot
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Quantiles Stats by Year')
sns.lineplot(ax=axes[0, 0], data=df, x= 'Year', y='Q1', color = "green")
sns.lineplot(ax=axes[0, 1], data=df, x= 'Year', y='Q2', color = "blue")
sns.lineplot(ax=axes[1, 0], data=df, x='Year', y='Q3', color = "yellow")
sns.lineplot(ax=axes[1, 1], data=df, x='Year', y='Q4', color = "black")


```
