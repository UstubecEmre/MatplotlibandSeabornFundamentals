# 1. What is a Rug Plot in Seaborn?

A rug plot is a simple plot that shows individual data points along a single axis (usually the x-axis) as small vertical or horizontal lines (like a "rug" lying on the axis).

# 2. Why Do We Use It

<table>
     <tr>
        <th>Purpose</th>
        <th>Explanation</th>
    </tr>
    <tr>
        <td>Show exact data points</td>
        <td>Useful for visualizing the actual position of every observation</td>
    </tr>
    <tr>
        <td>Complement KDE or histogram</td>
        <td>Helps verify where data is concentrated or sparse</td>
    </tr>
    <tr>
        <td>Understand sample size and clustering</td>
        <td>You can see if many values are stacked together in certain areas</td>
    </tr>
    <tr>
        <td>Good for small datasets</td>
        <td>With small to medium-sized data, it's easy to spot patterns</td>
    </tr>
</table>


### Note: When Not to Use Rug Plot:
> For very large datasets (thousands of points), rug plots can become unreadable.

> Too much overlap makes it hard to extract meaning.
