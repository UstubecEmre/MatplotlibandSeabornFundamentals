# Create Figure

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(10)
print(x)
```

```python
# calculate y and z values
y = x ** 2 + x +1
print("Calculated  Y Values:",y, '\n')

z = x ** 3 - x +2
print("Calculated Z Values:", z, "\n")
```

```python
# fig, ax = plt.subplots(2, 2)
figure = plt.figure()
ax1 = figure.add_axes([0.2, 0.2, 0.8, 0.8]) # left, bottom, width, height
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('x ** 2 + x - 0.02')
ax1.set_title('y =  x ** 2 + x - 0.02')


ax2 = figure.add_axes([0.30, 0.65, 0.3, 0.3]) # left, bottom, width, height
ax2.plot(x, z, 'g')
ax2.set_xlabel('x')
ax2.set_ylabel('x ** 3 - x + 0.01')
ax2.set_title('y = x ** 3 - x + 0.01')
```

```python
# examples
x = np.linspace(1, 100, 20)
y = x + 5
z = x + 15
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 6))
ax[0].plot(x, y, 'black')
ax[0].set_title("y = x + 5")
ax[0].set_xlabel("x")
ax[0].set_ylabel("x + 5")

ax[1].plot(x, z, 'maroon')
ax[1].set_title("z = x + 15")
ax[1].set_xlabel("x")
ax[1].set_ylabel("x + 15")

# if you want to save it, you can use this codes (fig.savefig('figure1.png'))
plt.tight_layout()
plt.show()
```
