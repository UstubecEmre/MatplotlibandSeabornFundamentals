# Matplotlib Library
## 1. What is Matplotlib?

Matplotlib is a popular Python library used to visualize data with graphs and charts.
It helps us turn numbers into picturesâ€”so we can see patterns, trends, and outliers more easily.

ðŸ”Ž Think of Matplotlib as Pythonâ€™s drawing tool for data.

## 2. How to import it
import matplotlib.pyplot as plt

> Note: We often use pyplot because it offers a simple, MATLAB-like interface.

## 3. Basic Terms You Should Know

1. Figure:

o The whole image or canvas you are working on.

o It can contain one or more plots.

> fig = plt.figure()

2. Axes:

o The actual plot or graph.

o Itâ€™s where your data is shown â€” like the X and Y axis in a math class.

o One figure can have multiple axes (charts).

> fig, ax = plt.subplots()

3. Subplot

o A smaller chart inside the main figure.

o You use subplots when you want to show multiple charts in the same figure.

> plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st chart


4. Plot

o The line or bar or scatter that shows your data.

o You use plot() to draw lines or points.

> plt.plot([1, 2, 3], [4, 5, 6])  # (x, y)

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a list

x = np.random.randint(1,50, 10)
plt.plot(x)
plt.title("Basic Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
y = [(i * 1.5 - 10) for i in x]
plt.plot(x, y, color = "skyblue", linewidth = 2, marker = "x")
plt.axis([0, 50, 0, 75]) # [x_min, x_max, y_min, y_max]
plt.title('Y Graph')
plt.xlabel("X Values")
plt.ylabel("y = x * 1.5 - 10")
plt.show() 
```

```python
x = np.linspace(0, 4, 12)
plt.plot(x, x, label = 'x')
plt.plot(x, x ** 0.25, label = 'x ^ (1 /4)')
plt.plot(x, x ** 3, label = 'x ^ 3')
plt.plot(x, x ** 0.5, label = 'x ^ (1 / 2)')
plt.plot(x, x + 15, label = 'x + 15')
plt.title("Examples of Charts")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# using subplot
fig, ax = plt.subplots(2)
ax[0].plot(x, x ** 2 - 24, color = 'olive')
ax[0].set_title('y = x ** 2 - 24')
ax[0].set_xlabel('x')
ax[0].set_ylabel('x ** 2 - 24')
ax[0].legend()

ax[1].plot(x, (x + 50) ** 0.5, color = 'darkslategray')
ax[1].set_title('y = (x +50) ** 0.5')
ax[1].set_xlabel('x')
ax[1].set_ylabel('(x +50) ** 0.5 ')
ax[1].legend()
plt.tight_layout() # Prevent overlapping
plt.show()
```

```python
# new example
x = np.random.randint(1, 20, 20)
fig, ax = plt.subplots(3, 2)

ax[0, 0].plot(x, x + 5, color = "olivedrab")
ax[0, 0].set_title("y = x + 5")
ax[0, 0].set_xlabel("x")
ax[0, 0].set_ylabel("y")


ax[0, 1].plot(x, x - 5, color = "seagreen")
ax[0, 1].set_title("y = x - 5")
ax[0, 1].set_xlabel("x")
ax[0, 1].set_ylabel("y")


ax[1, 0].plot(x, x + x ** 0.5, color = 'darkolivegreen')
ax[1, 0].set_title("y = x + x ** 0.5")
ax[1, 0].set_xlabel("x")
ax[1, 0].set_ylabel("y")


ax[1, 1].plot(x, x - x ** 0.5, color = "forestgreen")
ax[1, 1].set_title("y = x - x ** 0.5")
ax[1, 1].set_xlabel("x")
ax[1, 1].set_ylabel("y")


ax[2, 0].plot(x, x + 10, color = "darkseagreen")
ax[2, 0].set_title("y = x + 10")
ax[2, 0].set_xlabel("x")
ax[2, 0].set_ylabel("y")


ax[2, 1].plot(x, x - 10, color = "chartreuse")
ax[2, 1].set_title("y = x - 10")
ax[2, 1].set_xlabel("x")
ax[2, 1].set_ylabel("y")
plt.tight_layout()
plt.show()
```
