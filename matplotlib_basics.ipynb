## 3. Basic Terms You Should Know

1. Figure:

o The whole image or canvas you are working on.

o It can contain one or more plots.

> fig = plt.figure()

2. Axes:

o The actual plot or graph.

o Itâ€™s where your data is shown â€” like the X and Y axis in a math class.

o One figure can have multiple axes (charts).

> fig, ax = plt.subplots()

3. Subplot

o A smaller chart inside the main figure.

o You use subplots when you want to show multiple charts in the same figure.

> plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st chart


4. Plot

o The line or bar or scatter that shows your data.

o You use plot() to draw lines or points.

> plt.plot([1, 2, 3], [4, 5, 6])  # (x, y)

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a list

x = np.random.randint(1,50, 10)
plt.plot(x)
plt.title("Basic Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
y = [(i * 1.5 - 10) for i in x]
plt.plot(x, y, color = "skyblue", linewidth = 2, marker = "x")
plt.title('x * 1.5 - 10 Graph')
plt.xlabel("X Values")
plt.ylabel("y = x * 1.5 - 10")
plt.show() 
```
