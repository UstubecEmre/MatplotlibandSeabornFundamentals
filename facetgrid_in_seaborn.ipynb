# 1.What is FacetGrid in Seaborn?
A FacetGrid is a multi-plot grid — a layout that lets you split your dataset into subsets (called facets) and visualize each subset as a separate plot in the same figure.

✅ It's a powerful tool for:

Comparing relationships across categories

Automatically managing axes, titles, and layout

Creating rows × columns of plots based on categorical variables


# 2. Why Do We Use FacetGrid?
Because:

It lets you explore multi-dimensional relationships

You can break down your data by groups (e.g., gender, species, region)

It gives you a clear visual comparison of trends, distributions, or patterns



```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
file_path = r"D:\DataVisualizationForPython\datasets\orders.xlsx"

file = pd.read_excel(file_path)

if file is None:
    raise FileNotFoundError("File Path Not Found. Please, Check Your File Path")

df = pd.DataFrame(file)
df.head()
```

```python
# convert to lower case all columns
df.columns = df.columns.str.lower()
df.columns
```

```python
# explore some columns

# outlet_size
print("Unique Number of Outlet Size Column:",df['outlet_size'].nunique(),'\n')
print("Unique Values of Outlet Size Column\n:",df['outlet_size'].unique(),'\n')

print("*" *100)

# outlet_location_type
print("Unique Number of Outlet Location Type Column:",df['outlet_location_type'].nunique(),'\n')
print("Unique Values of Outlet Location Column\n:",df['outlet_location_type'].unique(),'\n')
```

```python
plt.figure(figsize = (20, 15))
facetgrid = sns.FacetGrid(data = df, col = "outlet_size")
facetgrid.map_dataframe(sns.histplot, x = "sales", palette = "coolwarm")
```

```python
facetgrid = sns.FacetGrid(data = df, col = "outlet_size")
facetgrid.map_dataframe(sns.lineplot, y = "sales", x = "outlet_year",palette = "coolwarm", ci = None)
```

```python
facetgrid = sns.FacetGrid(data = df, col = "outlet_size")
facetgrid.map_dataframe(sns.boxplot, x = "sales",palette = "coolwarm")
```

```python
facetgrid = sns.FacetGrid(data = df, col = "outlet_size")
facetgrid.map_dataframe(sns.kdeplot, x = "sales",color = "red")
```

```python
plt.figure(figsize= (25, 20))
facetgrid2 = sns.FacetGrid(data = df, col = "outlet_size", row = "outlet_location_type")#, sharey= False)
facetgrid2.map_dataframe(sns.histplot, x = "sales")
facetgrid2.set_axis_labels("sales","count")
facetgrid2.set_titles(col_template= '{col_name} Size', row_template= '{row_name} Type')
plt.tight_layout()
plt.show()
```

```python
# Lineplot: sales over outlet_year, separated by tier
plt.figure(figsize = (12, 8))
fg = sns.FacetGrid(df, col="tier", col_wrap= 2, height=3, aspect= 1)
fg.map(sns.lineplot, "outlet_year", "sales", marker ="D", color="green", linestyle = "--", ci = None)
fg.set_titles("Tier {col_name}")
fg.set_axis_labels("Year", "Sales")
fg.add_legend()
plt.tight_layout()
plt.show()
```

```python
# Histogram Example : Distribution of sales per tier
fg2 = sns.FacetGrid(df, col="tier", col_wrap=3, height = 5)
fg2.map(sns.histplot, "sales", bins=10, kde=True, color="forestgreen")
fg2.set_titles("Tier {col_name}")
fg2.set_axis_labels("Sales", "Count")
plt.tight_layout()
plt.show()
```

```python
# Boxplot Example:  Sales by outlet_year, faceted by tier

fg3 = sns.FacetGrid(df, col="tier", height= 4, sharex= False, sharey= False)
fg3.map_dataframe(sns.boxplot, x="outlet_year", y="sales", palette="Set2")
fg3.figure.suptitle("Sales By Outlet Year (Faceted By Tier)", fontsize = 16)
fg3.figure.supxlabel("Year")
fg3.figure.supylabel("Total Sales")
fg3.figure.set_facecolor(color= "beige")
fg3.figure.set_figwidth(15)
fg3.figure.set_figheight(12)

fg3.set_titles("Tier {col_name}")
fg3.set_xticklabels(rotation=45)
plt.tight_layout()
plt.show()
```
