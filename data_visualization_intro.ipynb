# 1. üìä What is Data Visualization in Python?

Data Visualization is the graphical representation of information and data. In Python, it means using libraries like Matplotlib, Seaborn, Plotly, and Pandas to turn raw data into meaningful visual forms such as:

> Bar charts

> Line plots

> Histograms

> Scatter plots

> Heatmaps

> Pie charts

> Interactive dashboards

These visualizations help reveal patterns, trends, and outliers in data ‚Äî things that would be difficult or even impossible to see in raw tables or spreadsheets.

```python
# import libraries
import numpy as np # linear algebra
import matplotlib.pyplot as plt # data visualization
```

```python
# Bar plot example

# create numeric samples
x = list(range(0, 11))
y = [i ** 2 for i in x]

# visualize
plt.plot(x, y, 'go--', linewidth = 2, markersize = 4 )
plt.axis([0, 12, 0, 120])
plt.title('Example of Bar Chart')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Examples
x = np.linspace(0,5,50)
plt.plot(x, x, label = 'linear', color = 'purple')
plt.plot(x, x ** 2, label = 'quadratic', color = 'maroon')
plt.plot(x, x ** 3, label = 'cubic', color = "magenta")
plt.plot(x, x ** 0.5, label = 'squareroot', color = "lightpink")
plt.xlabel("Data")
plt.ylabel("Processed Data")
plt.title("Chart")
plt.legend()
```

```python
# create a dataset
year = [2019, 2020, 2021, 2022]

goals_of_messi = [44, 27, 43, 11]
goals_of_lewandowski = [47, 47, 69, 50]
goals_of_ronaldo = [32, 44, 46, 35]

plt.plot([],[], color = "darkkhaki", label = "goals_of_messi")
plt.plot([],[], color = "olive", label = "goals_of_lewandowski")
plt.plot([],[], color = "darkolivegreen", label = "goals_of_ronaldo")

plt.stackplot(year, goals_of_messi, goals_of_lewandowski, goals_of_ronaldo, colors = ["darkkhaki","olive","darkolivegreen"])
plt.title("Goals Scored By Years")
plt.xlabel("Year")
plt.ylabel("Number of Goals")
plt.legend()
plt.show()
```
